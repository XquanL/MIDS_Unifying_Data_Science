## Exercise 1

```python
cpe = pd.read_csv("https://media.githubusercontent.com/media/nickeubank/MIDS_Data/master/"+
"descriptive_exercise/crime_expend_MA.csv")
cpe.sample(10)
```

## Execise 2

```python
# select county with county_code 4 and 10
cpe_4_10 = cpe[(cpe.county_code == 4) | (cpe.county_code == 10)]
# calculate the mean expenditure level and mean crimeindex score
# county_code 4
cpe_4 = cpe[cpe.county_code == 4]
cpe_4_expenditure_mean = round(cpe_4.policeexpenditures.mean(), 2)
cpe_4_crimeindex_mean = round(cpe_4.crimeindex.mean(),2)
print(f"for county 4, average policing expenditure is {cpe_4_expenditure_mean}",
f"and average crime index is {cpe_4_crimeindex_mean}")
# county_code 10
cpe_10 = cpe[cpe.county_code == 10]
cpe_10_expenditure_mean = round(cpe_10.policeexpenditures.mean(), 2)
cpe_10_crimeindex_mean = round(cpe_10.crimeindex.mean(),2)
print(f"for county 10, average policing expenditure is {cpe_10_expenditure_mean}",
f" and average crime index is {cpe_10_crimeindex_mean}")
```

```python
# calculate the mean expenditure level and mean crimeindex score by loop
for county_code in [4, 10]:
    cpe_county = cpe[cpe.county_code == county_code]
    cpe_county_expenditure_mean = round(cpe_county.policeexpenditures.mean(), 2)
    cpe_county_crimeindex_mean = round(cpe_county.crimeindex.mean(),2)
    print(f"for county {county_code}, average policing expenditure is {cpe_county_expenditure_mean}",
    f"and average crime index is {cpe_county_crimeindex_mean}")
```

```python
# calculate the standard deviation of both expenditures and crime for these two counties by loop
for county_code in [4, 10]:
    cpe_county = cpe[cpe.county_code == county_code]
    cpe_county_expenditure_std = round(cpe_county.policeexpenditures.std(), 2)
    cpe_county_crimeindex_std = round(cpe_county.crimeindex.std(),2)
    print(f"for county {county_code}, standard deviation of policing expenditure is",
    f"{cpe_county_expenditure_std} and standard deviation of crime index is {cpe_county_crimeindex_std}")
```

## Exercise 3

```python
# calculate the correlation of both expenditures and crime for these two counties
for county_code in [4,10]:
    cpe_county = cpe[cpe.county_code == county_code]
    cpe_county_expenditure_corr = round(cpe_county.policeexpenditures.corr(cpe_county.crimeindex), 4)
    print(f"for county {county_code}, ", 
    f"correlation of policing expenditure and crime index is {cpe_county_expenditure_corr}")
```

## Exercise 6

```python
#plot histograms of policeexpenditures for both county 4 
# and county 10 in the same figure with different colors
import matplotlib.pyplot as plt
plt.hist(cpe_4.policeexpenditures, color = "blue", alpha = 0.5, label = "county_4")
plt.hist(cpe_10.policeexpenditures, color = "red", alpha = 0.5, label = "county_10")
plt.legend(loc = "upper right")
plt.title("Histogram of Police Expenditures for County 4 and County 10 ")
plt.xlabel("Police Expenditures")
plt.ylabel("Frequency")
plt.show()
```

> Yes. Based on the hitogram above, county 4 and county 10 do not have a similar relationship between crime and police expenditures.

## Exercise 7

```python
# create a scatter plot of the relationship between crime and police expenditures for each county
for county_code in [4, 10]:
    cpe_county = cpe[cpe.county_code == county_code]
    plt.scatter(cpe_county.policeexpenditures, cpe_county.crimeindex, label = f"county_{county_code}")

plt.legend(loc = "upper right")
plt.title("Scatter Plot of Crime and Police Expenditures for County 4 and County 10 ")
plt.xlabel("Police Expenditures")
plt.ylabel("Crime Index")
plt.show()
```

> Based on the scatter plot above, the relationship between crime and police expenditures for county 4 and county 10 are not similar at all.
